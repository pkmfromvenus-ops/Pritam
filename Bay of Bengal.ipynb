```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
ds = xr.open_dataset(r"C:\Users\DELL\Desktop\28.01.26\sst.day.mean.2020.nc")
```

```python
sst = ds["sst"]
```

```python
sst_basin = sst.sel(lon=slice(80,100), lat=slice(5,25))
```

```python
sst_mam = sst_basin.sel(time=sst_basin.time.dt.month.isin([3,4,5]))
sst_mam_mean = sst_mam.mean(dim="time")
```

```python
sst_jjas = sst_basin.sel(time=sst_basin.time.dt.month.isin([6,7,8,9]))
sst_jjas_mean = sst_jjas.mean(dim="time")
```

```python
sst_on = sst_basin.sel(time=sst_basin.time.dt.month.isin([10,11]))
sst_on_mean = sst_on.mean(dim="time")
```

```python
vmin_val = 25
vmax_val = 31
```

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.PlateCarree())

sst_mam_mean.plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    cmap="YlOrRd",
    vmin=vmin_val,
    vmax=vmax_val,
    cbar_kwargs={"label": "SST (C)"}
)

ax.coastlines()
ax.set_title("Pre-Monsoon SST (MAM) 2020")
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.PlateCarree())

sst_jjas_mean.plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    cmap="YlOrRd",
    vmin=vmin_val,
    vmax=vmax_val,
    cbar_kwargs={"label": "SST (C)"}
)

ax.coastlines()
ax.set_title("Monsoon SST (JJAS) 2020")
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.PlateCarree())

sst_on_mean.plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    cmap="YlOrRd",
    vmin=vmin_val,
    vmax=vmax_val,
    cbar_kwargs={"label": "SST (C)"}
)

ax.coastlines()
ax.set_title("Post-Monsoon SST (ON) 2020")
plt.show()
```
